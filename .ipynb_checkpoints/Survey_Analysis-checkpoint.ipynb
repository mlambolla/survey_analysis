# Convertir a categóricas, Negativo, Neutral, Positivo

```python
conversion = {1:'Negativo',2:'Negativo',3:'Neutral',4:'Positivo',5:'Positivo','NaN':'Sin Opinion'}
conversion_preg5 = {5:'Negativo',4:'Negativo',3:'Neutral',1:'Positivo',2:'Positivo'}
```

```python
#2_elementos_contra_accidentes	3_dificultad_balance_trabajo_personal	4_significado_empresa	5_no_me_siento_familia_empresa	6_recomiendo_amigo
df['1_capacitacion_cat'] = df['1_capacitacion'].map(lambda x: x if pd.isnull(x) else conversion[x])
df['2_elementos_contra_accidentes_cat'] = df['2_elementos_contra_accidentes'].map(lambda x: x if pd.isnull(x) else conversion[x] )
df['3_dificultad_balance_trabajo_personal_cat'] = df['3_dificultad_balance_trabajo_personal'].map(lambda x: x if pd.isnull(x) else conversion[x])
df['4_significado_empresa_cat'] = df['4_significado_empresa'].map(lambda x: x if pd.isnull(x) else conversion[x])
df['5_no_me_siento_familia_empresa_cat'] = df['5_no_me_siento_familia_empresa'].map(lambda x: x if pd.isnull(x) else conversion_preg5[x])
df['6_recomiendo_amigo_cat'] = df['6_recomiendo_amigo'].map(lambda x: x if pd.isnull(x) else conversion[x])
```
